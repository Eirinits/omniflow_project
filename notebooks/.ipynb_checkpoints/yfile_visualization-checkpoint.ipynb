```python
from yfiles_jupyter_graphs import GraphWidget
from typing import Dict
import pandas as pd
```

importing edges dataframe (normally it should be complemented with a dataframe containing the info on the nodes)

```python
df=pd.read_csv("edges_test.csv",index_col=0).reset_index(drop=True)
df
```

```python
df_nodes=pd.read_csv("nodes_test.csv",index_col=0)
df_nodes
```

```python
# creating empty object for visualization
w = GraphWidget()
```

```python
# filling w with nodes
objects = []
for idx, item in df_nodes.iterrows():
    obj = {
        "id" : df_nodes["Genesymbol"].loc[idx],
        "properties" : {"label": df_nodes["Genesymbol"].loc[idx]},
        "color":"#ffffff",
        "styles":{"backgroundColor":"#ffffff"}
        }
    objects.append(obj)
w.nodes=objects
```

```python
# filling w with edges
objects= []
for index, row in df.iterrows():
    obj={
    "id":df["Effect"].loc[index],
    "start" : df["source"].loc[index],
    "end" : df["target"].loc[index],
    "properties":{"references":df["References"].loc[index]}}
    objects.append(obj)
w.edges=objects
```

```python
def custom_edge_color_mapping(edge: Dict):
    """let the edge be purple if the starting node has an even index"""
    return ("#ff0066" if edge['id']  == "inhibition" else "#00cc00")
w.set_edge_color_mapping(custom_edge_color_mapping)


def custom_label_styles_mapping(node: Dict):
    """let the label be the negated purple big index"""
    return {
        'backgroundColor': '#ffffff', 
        'color': '#ffffff', 
        'shape':'round-rectangle'
    }
w.set_node_styles_mapping(custom_label_styles_mapping)
```

```python
w.directed=True
w.graph_layout='hierarchic'
w.show()
```
