```python
%%time
from neko.core.network import Network
from neko._visual.visualize_network import NetworkVisualizer
from neko._inputs.resources import Resources
from neko._annotations.gene_ontology import Ontology
from neko._methods.methods_update import Connections
import omnipath as op
import pandas as pd
```

# Getting the resources

## SIGNOR

```python
resources = Resources()
resources.import_signor_tsv("all_data_31_05_24.tsv") 
```

```python
initial_nodes = [
"PIK3CA",
"MAP2K1",
"GSK3B",
"MAPK14",
"CTNNB1",
"AKT1",
"MAP3K7"]
```

```python
output_nodes_antisurvival = [
    'FOXO3',
    'CASP8',
    'CASP9'
]

output_nodes_prosurvival = [
    'CCND1',
    'MYC',
    'TCF7L2'
]
```

# Using as resources just Signor

```python
new_net = Network(initial_nodes=initial_nodes, resources=resources.interactions)
```

```python
%%time
new_net.connect_as_atopo(max_len=3, strategy="complete", outputs=output_nodes_antisurvival)
```

```python
new_net.nodes
```

```python
visualizer = NetworkVisualizer(new_net, color_by='effect', noi=True)
visualizer.render()
```

```python
visualizer = NetworkVisualizer(new_net, color_by='effect', noi=True, predefined_node="FOXO3")
visualizer.render()
```

```python
visualizer = NetworkVisualizer(new_net, color_by='effect', noi=True, predefined_node="CASP9")
visualizer.render()
```

# Putting toghether Pro and Anti survival nodes using SIGNOR

```python
full_net = Network(initial_nodes=initial_nodes, resources=resources.interactions)
```

```python
%%time
outputs = output_nodes_antisurvival + output_nodes_prosurvival
full_net.connect_as_atopo(max_len=3, strategy="complete", outputs=outputs)
```

```python
full_net.edges
```

```python
visualizer = NetworkVisualizer(full_net, color_by='effect', noi=True)
visualizer.render()
```
