```python
%%time
from omniflow_project.core.network import Network
from omniflow_project._visual.visualize_network import NetworkVisualizer
from omniflow_project._inputs.resources import Resources
from omniflow_project._annotations.gene_ontology import Ontology
import omnipath as op
```

### First we define a set of node we are interested in

```python
genes = ["SRC", "NOTCH1", "FAK", "CDH1", "CDH2", "VIM", "MAP4K4", "LATS1", "LATS2", "PTK2B"]
```

### Next we create a network using as input the gene set. Note that if no resource is specified, the default used resource by the package is omnipath.all_interactions()

```python
new_net1 = Network(genes)
```

```python
visualizer = NetworkVisualizer(new_net1, color_by='effect')
visualizer.render()
```

### Now we can explore if the nodes are connected without introducing new intermidiate nodes

```python
%%time
new_net1.connect_nodes(only_signed=True, consensus_only=True)
```

```python
new_net1.nodes
```

### We can visualize the network with the following command. The network picture is also exported in the working repository, in pdf format

```python
visualizer = NetworkVisualizer(new_net1, color_by='effect')
visualizer.render()
```

### We can try to connect the disconnected components using the following function. Be careful that the parameter maxlen is used to verify that in the network all the nodes have a distance < than maxlen. The parameter only_signed forces the algorithm to look for just signed interactions. Consensus retrieve those interactions with consensus between the references. Finally, connect_node_when_first_introduced, changes the base algorithm, making it look for possible connection between all the disconnected node in the network when a new bridge-gene is introduced in the network.

```python
%%time
new_net1.complete_connection(mode="ALL", maxlen=6, k_mean="tight", only_signed=True, connect_node_when_first_introduced=True, consensus=True)
```

```python
visualizer1 = NetworkVisualizer(new_net1, color_by='effect')
visualizer1.render(output_file='network2')
```

```python
annot = Ontology()
#user_defined_tissue = 'breast cancer'
results_df = annot.check_tissue_annotations(genes_df = new_net1.nodes, tissue = 'colorectal cancer')
print(results_df)
```

```python
network_visualizer = NetworkVisualizer(new_net1)
network_visualizer.tissue_mapping(results_df)
network_visualizer.render()
```

```python
visualizer = NetworkVisualizer(new_net1, predefined_node="EGFR",color_by='effect')
visualizer.render()
```

### We can access the network object edges to explore which interactions are in our network

```python
#We can access to the edges of the network
new_net1.edges
```

# If the all_interaction database of omnipath is not sufficient, we can change it or extend it. To do it, we can use the Resources object:

```python
interactions = op.interactions.AllInteractions.get() #first we get the interactions from omnipath
```

### Then, we can add allInteractions in the Resources object

```python
resources = Resources()
resources.add_database(interactions)
```

```python
resources.interactions
```

### Now we can explore another database we are interested in, for example, the ligand receptor one

```python
extra = op.interactions.LigRecExtra.get()
```

### And then we add it to the Resources object

```python
resources.add_database(extra, reset_index=True)
```

### Now we create the network with a similar list of genes but specifying the Resource object

```python
genes = ["SRC", "NOTCH1", "PTK2", "CDH1", "CDH2", "VIM", "MAP4K4", "PTK2B", "COL1A1"]
```

```python
%%time
new_net1 = Network(genes, resources=resources.interactions)
```

```python
%%time
new_net1.connect_nodes(only_signed=True, consensus_only=True)
```

```python
%%time
new_net1.complete_connection(mode="ALL", maxlen=6, k_mean="tight", only_signed=True,
                            consensus=True, connect_node_when_first_introduced=True)
```

```python
visualizer = NetworkVisualizer(new_net1, color_by='effect')
visualizer.render()
```

## The whole tool, as the name suggest, is based on Omnipath. But the user can also integrate its own databases. As example, we show how to integrate the Signor database. In order to do so, the User needs to have already downloaded the whole Signor database, available at the following link: https://signor.uniroma2.it/downloads.php 
## Omniflow has been conceived to be a modular tool, with the possibility to add new databases. Every time a User wants to add a database, it should prepare an apposite function in the Resources module, in order to convert the target database, in the Omniflow format. We suggest to use as example, the function created for the Signor database. In future releases we plan to add more function to include more databases.

```python
resources = Resources()
resources.import_signor_tsv("./all_data_18_03_24.tsv") # this function accept only tab separated values
```

```python
%%time
new_net1 = Network(genes, resources=resources.interactions) # here we load the signor interactions
```

## It is important to note, that not all the databases have the same structure as Omnipath. In particular, if the columns "consensus" are missing, we suggest to avoid to use the flag "consensus" or "consensus_only" when using Omniflow with customize databases.

```python
%%time
new_net1.complete_connection(mode="ALL", maxlen=6, k_mean="tight", only_signed=True,
                             connect_node_when_first_introduced=True) # here, consensus_only is missing, in this way it is automatically False
```
